1.) Create an ImageAnalyzer class which is able to:
- read an image (e.g. a photo in .jpg format)
- plot the image
- get the image size (numer pf pixels in rows and columns)
- get values of single pixels
- plot a gray-scale version of the image
- create a single plot showing overlapping histograms of RGB colors
- write the RGB-color values to separate columns in a pandas data frame 

```python
class ImageAnalyzer:

    def readImage(self, path):
        return plt.imread(path)
    
    def showImage(self, image):
        plt.imshow(image)
        plt.axis('off')  # Hide the axes for a cleaner look
        plt.show()

    def getImageSize(self, image):
        return image.shape
    
    def valueOfPixel(self, image, x, y):
        return image[x][y]
    
    def showGrayScaledImage(self, image):
        image = np.mean(image, axis=2)

        plt.imshow(image, cmap='gray')
        plt.axis('off') 
        plt.show()
    
    def showSinglePlotWithOverlappingHistograms(self, image):
        plt.hist(image.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
        plt.show()

    def getRGBColorsAsDataFrame(self, image):
        return pd.DataFrame(image.reshape(-1, 3), columns=['R', 'G', 'B'])
```
